```python
import numpy as np 
import matplotlib.pyplot as plt
import scienceplots
from scipy.special import lambertw 
```

```python
def f1(alpha, K, B):
    return K * np.exp(-B*alpha)
```

```python
bc = - lambertw(-1/(2*np.exp(1)), -1).real 
alpha_c = 1-1/bc
```

```python
# Using a standardized style set that largely holds to Nature's guidelines.
plt.style.use('science')
plt.style.use(['science','nature'])
```

```python
width = 188/25.4
fig = plt.figure(figsize=(width, width/3))

import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1.5])  # The width_ratios parameter helps in specifying the relative widths of the columns

# Create the first subplot (figure) that spans 1 column
ax1 = fig.add_subplot(gs[0])

# Create the second subplot (figure) that spans 2 columns
ax2 = fig.add_subplot(gs[1])

# fig 2
K = 2
alphas = np.linspace(0, 1.2, 1000)
ax2.plot(alphas, 1-alphas, label=r'$y=1-\alpha$')
ax2.plot(alphas, f1(alphas, K, 3.5), label=r'$y=f(\alpha)$ for $B=3.5$' )
ax2.plot(alphas, f1(alphas, K, 2.67835), color='red', lw=2, label=r'$y=f(\alpha)$ for $B=B_{c}^*\approx 2.678$')
ax2.plot(alphas, f1(alphas, K, 2), label=r'$y=f(\alpha)$ for $B=2$')

ax2.set_ylim(0, 1.2)
ax2.legend(frameon=False)
ax2.set_xlabel(r'$\alpha$')

# fig 1
x = np.linspace(-1/np.exp(1)+1e-4,0, 1000)
ax1.plot(x, lambertw(x, 0).real, label=r'$W_{0}(x)$')
ax1.plot(x, lambertw(x, -1).real, label=r'$W_{-1}(x)$')
ax1.set_xlabel(r'$x$')
ax1.legend(frameon=False)
fig.tight_layout()
fig.savefig("lambertw.pdf")
```
