```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
no_sparsity_V= np.load("data/no_sparsity.npy")
sparsity_10_V = np.load("data/sparse.npy")
sparsity_50_V = np.load("data/sparse_50.npy")
sparsity_30_V = np.load("data/sparse_30.npy")
sparsity_01_V = np.load("data/sparse_01.npy")
```

```python
buffer_sizes_1lvl = np.arange(0,9,0.05)
buf_x=np.arange(3.4,3.7,0.01)
buffersizes_tot = np.concatenate((buffer_sizes_1lvl, buf_x))
buffer_sizes = np.arange(0,9,0.05)
```

```python
fig, ax1 = plt.subplots(figsize=(6,6))

size = 14

B1=3.6799119729375
B2=3.6837088479375
B3=3.6894041604375
B4=3.6979471291875
B5=3.7107615823125
B6 = 3.6000
B_values = [B1, B2, B3, B4, B5, B6]
norm = plt.Normalize(min(B_values[:-1])*0.995, max(B_values[:-1]))
colors = plt.cm.copper(norm(B_values))
colors[-1] = plt.cm.copper(1.0)
norm = matplotlib.colors.SymLogNorm(linthresh=0.01, vmin=min(B_values), vmax=max(B_values))



ax1.scatter(buffersizes_tot, no_sparsity_V, c=colors[0], label= r"$\mathrm{STN,}\ \mathrm{sparsity} = 0\%$", s=size, marker= "s")
ax1.scatter(buffersizes_tot, sparsity_50_V, label= r"$\mathrm{STN,}\ \mathrm{sparsity}  = 50\%$ ", c = colors[1], s=size+20, marker="+")
ax1.scatter(buffer_sizes, np.mean(sparsity_30_V,axis=1), label= r"$\mathrm{STN,}\ \mathrm{sparsity}  = 70\%$", c = colors[2], s=size+7, marker="^")
ax1.scatter(buffersizes_tot, sparsity_10_V, label= r"$\mathrm{STN,}\ \mathrm{sparsity}  = 90\%$", c = colors[3], s=size+7,marker="v")
ax1.scatter(buffersizes_tot, sparsity_01_V, c=colors[4], label= r"$\mathrm{STN,}\ \mathrm{sparsity}  = 99\%$", s=size+20, marker="*")
#plt.scatter(buffersizes_tot, taus_het_k, c="b", label= r"STN with heterogenous $k$", s=5)

ax1.set_xlim(-0.1, 6)
ax1.set_ylim(-0.3, 6)

ax1.set_xlabel(r"$B$", fontsize=16)
ax1.set_ylabel(r"$v$", rotation=0, labelpad=10, fontsize=18)
ax1.legend(frameon=False, fontsize= 14, loc="upper left", bbox_to_anchor=(-0.02, 1.02), handlelength=.5, handletextpad=0.4)
#plt.title("Effect of heterogeneity and sparsity on temporal network criticality.")
#plt.yscale("log")


left, bottom, width, height = [0.59, 0.26, 0.29, 0.558]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.scatter(buffersizes_tot, no_sparsity_V, c=colors[0], label= r"$STN, sparsity = $0\%$", s=size, marker= "s")
ax2.scatter(buffersizes_tot, sparsity_50_V, label= r"$STN, sparsity = $50\%$ ", c = colors[1], s=size+20, marker="+")
ax2.scatter(buffer_sizes, np.mean(sparsity_30_V,axis=1), label= r"$STN, sparsity = $70\%$ ", c = colors[2], s=size+7, marker="^")
ax2.scatter(buffersizes_tot, sparsity_10_V, label= r"$STN, sparsity = $90\%$ ", c = colors[3], s=size+7,marker="v")
ax2.scatter(buffersizes_tot, sparsity_01_V, c=colors[4], label= r"$STN, sparsity = 99\%$ ", s=size+20, marker="*")

ax2.set_yscale("log")
ax2.set_xscale("log")
ax2.set_xlabel(r"$B$", fontsize=16)
ax2.set_ylabel(r"$v$", rotation=0, labelpad=10, fontsize=18)
ax2.set_ylim(0.5E-9, 1E1)



bc = 3.674
ax2.vlines(bc, ymin=1E-10, ymax = 30, ls=(0,(3,1.5)), color="black", 
            linewidth=1.5)#label=r"$\mathrm{workplace}\ B_\mathrm{c} \approx$ " +r"$2.54$",
arrow_x = bc+0.07 # x-coordinate of the arrow
arrow_y = 2E-0 # y-coordinate of the arrow
text_x = .4E-1# x-coordinate of the text box
text_y = 23*1E-0 # y-coordinate of the text box
ax2.annotate(r'$B_\mathrm{c}(N) \approx 3.674$', 
            xy=(arrow_x, arrow_y), xytext=(text_x, text_y),arrowprops=dict(arrowstyle='->'), color='black', fontsize=16,
            zorder = 100)

ax2.yaxis.set_minor_locator(plt.LogLocator(base=10.0, subs=(1.0,)))
ax2.yaxis.set_tick_params(direction="in")
ax2.tick_params(axis="both", labelsize=16)
ax2.xaxis.set_minor_locator(plt.LogLocator(base=10.0, subs=(0.1,)))
ax2.xaxis.set_tick_params(direction="in")

ax1.text(0.995, .99, r"$\mathrm{(b)}$", transform=ax1.transAxes,
        horizontalalignment='right', verticalalignment='top', fontsize=20)

plt.savefig("sparsity_v_graph.pdf", bbox_inches="tight")
#plt.yscale("log")
```
